##### TASKS :
* This dataset provides a detailed overview of gym members' exercise routines, physical attributes, and fitness metrics. It contains 973 samples of gym data, including key performance indicators such as heart rate, calories burned, and workout duration. Each entry also includes demographic data and experience levels, allowing for comprehensive analysis of fitness patterns, athlete progression, and health trends.
###### FEATURES:
* Age: Age of the gym member.
* Gender: Gender of the gym member (Male or Female).
* Weight (kg): Member’s weight in kilograms.
* Height (m): Member’s height in meters.
* Max_BPM: Maximum heart rate (beats per minute) during workout sessions.
* Avg_BPM: Average heart rate during workout sessions.
* Resting_BPM: Heart rate at rest before workout.
* Session_Duration (hours): Duration of each workout session in hours.
* Calories_Burned: Total calories burned during each session.
* Workout_Type: Type of workout performed (e.g., Cardio, Strength, Yoga, HIIT).
* Fat_Percentage: Body fat percentage of the member.
* Water_Intake (liters): Daily water intake during workouts.
* Workout_Frequency (days/week): Number of workout sessions per week.
* Experience_Level: Level of experience, from beginner (1) to expert (3).
* BMI: Body Mass Index, calculated from height and weight.

```python
# Lets first check the distribution of age 
plt.title("HISTOGRAM: AGE FEATURE")
sns.histplot(data=df['Age'], kde=True, bins=20)
plt.show()
```

* From the analysis, we understood that most of the GYM customers are in the age range of (40-60) as compare to (20-40).

```python
# Lets understand the difference of fat in the body with the Years of exoerience.
plt.title("BAR PLOT: EXPERIENCE LEVEL VS FAT PERCENTAGE")
sns.barplot(data=df, x='Experience_Level', y='Fat_Percentage', estimator='median', hue='Gender',width=0.5,gap=0.1)
plt.show()
```

```python
# scatter plot : Calories burn and Average BPM and Session_Duration (hours)
fig, axes = plt.subplots(1,2,figsize=(8,5))

# Scatter  plot for calroes burned  with BPM
axes[0].set_title("Calories burn and AVG BPM")
sns.scatterplot(data=df, x='Calories_Burned', y='Avg_BPM',hue="Gender",markers=True, ax=axes[0])

# scatter plot for calories burned wirh session duration
axes[1].set_title("Calories burn and AVG BPM")
sns.scatterplot(data=df, x='Calories_Burned', y='Session_Duration (hours)',hue="Gender", ax=axes[1])

plt.tight_layout()
plt.show()
```

```python
# High corelation since, it is a direct relation
df['Session_Duration (hours)'].corr(df["Calories_Burned"])
```

```python
# Session duration is not a decision parameter for the BPM(Beats per minute )
df['Session_Duration (hours)'].corr(df['Avg_BPM'])
```

```python
# Group by analysis 
df.groupby("Workout_Type")['Calories_Burned'].mean() # HIIT(High intensity interval training) has requires to burn high amount of calories.
```
